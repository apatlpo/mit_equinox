### line deployment

Base segments are 40 nm long

```python
lon_north = [5.8, 5.6]
lat_north = [42.9, 42.3]

lon_south = [4.1, 4.3]
lat_south = [40.4, 41]
```

#### North

```python
lon, lat = lon_north, lat_north
bbox = bbox_north
```

```python
# quick util

def process(dl, speed=5):
    #dl = 4 # in nautical miles
    speed = 5 # knots
    lon_i, lat_i, time = ca.follow_path(lon, lat, dl, speed)

    fig, ax = ca.plot_site(bbox, figsize=(5,5))
    ax.plot(lon_i, lat_i, '-+', transform=ccrs.PlateCarree())
    #print(" ".join(["{:.1f}".format(dl, t) for t in time]) )
```

```python
for dl in [4, 8, 16]:
    process(dl)
```

#### South

```python
lon, lat = lon_south, lat_south
bbox = bbox_south
```

```python
for dl in [4, 8, 16]:
    process(dl)
```

#### Back and forth in the south

```python
lon, lat = lon_south, lat_south

lon = lon + [lon[0]]
lat = lat + [lat[0]]

bbox = bbox_south
```

```python
for dl in [4, 8, 16]:
    process(dl)
```
