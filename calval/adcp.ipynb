```python
import os
from glob import glob

import numpy as np
import pandas as pd
import xarray as xr

%matplotlib inline
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.geodesic as cgeo
crs = ccrs.PlateCarree()
import cmocean.cm as cm

import warnings
warnings.filterwarnings("ignore")
```

```python
med_dir = "/home/datawork-lops-oc/MED/final_dataset"

campaign = "protevsmed_2017_leg1"

campaign_dir = os.path.join(med_dir, campaign)

def load_adcp(path, average="LTA", freq=150):
    file_path = os.path.join(path, "vmadcp_"+average+f"_{freq}/*.nc")
    files = glob(file_path)
    assert len(files)==1, f"Multiple files at {file_path}"
    ds = xr.open_dataset(files[0])
    return ds

ds = load_adcp(campaign_dir)
```
